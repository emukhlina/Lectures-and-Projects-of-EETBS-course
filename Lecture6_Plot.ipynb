```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(-10,10.1,0.1) #start point, end point which is not included and step
```

```python
y=x**2
```

```python
plt.figure()
plt.grid()
plt.plot(x,y)
```

```python
fig1 = plt.figure()
plt.plot(x, np.sin(x), "-")
plt.plot(x,np.cos(x),"--")
plt.grid()
fig1.savefig("MyArt.png")
```

## If we prefer subplots

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(x, np.sin(x))
plt.subplot(2,1,2)
plt.plot(x, np.cos(x))
```

## If we dont like 2,1 approach, we use another

```python
fig, ax = plt.subplots(2)
```

```python
ax[1].plot(x,np.sin(x))
ax[0].plot(x,np.cos(x))
fig
```

## Colors

```python
#can use name and shortcut of the color and grayscale and Hex code and RGB and HTML colors
plt.figure()
plt.plot(x, np.sin(x), color="blue")
plt.plot(x, np.sin(x-1), color="g")
plt.plot(x, np.sin(x-2), color="0.75")
plt.plot(x, np.sin(x-3), color="#FFDD44")
plt.plot(x, np.sin(x-4), color=(1,0.2,0.3))
plt.plot(x, np.sin(x-5), color="chartreuse")
```

# if we wanna change style of lines

```python
plt.figure()
plt.plot(x, x, linestyle="solid")
plt.plot(x, x+1, linestyle="dashed")
plt.plot(x, x+2, linestyle="dotted")
plt.plot(x, x+3, linestyle="dashdot")
```

# shortcuts also for linestyles

```python
plt.figure()
plt.plot(x, x, linestyle="-")
plt.plot(x, x+1, linestyle="--")
plt.plot(x, x+2, linestyle=":")
plt.plot(x, x+3, linestyle="-.")
```

```python
plt.figure()
plt.plot(x, x, "-g")
plt.plot(x, x+1, "--c")
plt.plot(x, x+2, ":r")
plt.plot(x, x+3, "-.k")
```

### If we wanna configure the axis

```python
plt.figure()
plt.plot(x, x, linestyle="-")
plt.plot(x, x+1, linestyle="--")
plt.plot(x, x+2, linestyle=":")
plt.plot(x, x+3, linestyle="-.")
plt.xlim(2,5)
plt.ylim(2,6)
```

```python
plt.figure()
plt.plot(x, x, linestyle="-")
plt.plot(x, x+1, linestyle="--")
plt.plot(x, x+2, linestyle=":")
plt.plot(x, x+3, linestyle="-.")
plt.xlim(2,5)
plt.ylim(2,6)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("fancy title")
```

```python
plt.figure()
plt.plot(x, np.cos(x), "o", color="red")
```

## Scatter Plot

```python
plt.figure()
plt.scatter(x, np.sin(x), marker="o")
```

## Let's do something more fancy

```python
import os
```

```python
rng=np.random.RandomState(80)
x=rng.randn(500)
y=rng.randn(500)
randomcolors=rng.randn(500)
randomsizes = 1000*rng.randn(500)
fig7=plt.figure(figsize=(15,8),dpi=400)
plt.scatter(x,y,c=randomcolors,s=randomsizes,alpha=0.5,cmap="viridis")
plt.colorbar()
```
