```python
name_joinedFile_WithLaggedAndTimeRelatedFeatures= "pretty_joined_DF_WtihLaggedAndTimeRelatedFeature.csv"
path_processedDataFolder = "/content/drive/MyDrive/EETBS 2023-2024/ProcessedData"
path_joinedFile_WithLaggedAndTimeRelatedFeatures = os.path.join(path_processedDataFolder,name_joinedFile_WithLaggedAndTimeRelatedFeatures)
DF_joined=pd.read_csv(path_joinedFile_WithLaggedAndTimeRelatedFeatures,index_col=0, parse_dates=[0])
DF_joined.dropna(inplace=True)
DF_joined
```

```python
DF_training=DF_joined.loc["2014-03-01":"2014-08-30"]
DF_training_target=DF_training.loc[:,"consumption"]
DF_training_features =DF_training.drop(["consumption","temperature","irradiance","weekOfYear"],axis=1)
DF_training_features
```

```python
DF_validation=DF_joined.loc["2014-08-31":"2014-09-14"]
DF_trainingandValidation=DF_joined.loc["2014-03-01":"2014-09-15"]
DF_test = DF_joined.loc["2014-09-15":"2014-09-30"]
```

Linear Regression

```python
from sklearn import linear_model
linear_reg=linear_model.LinearRegression()
linear_reg.fit(DF_training_features, DF_training_target)
```

```python
DF_validation_target = DF_validation.loc[:,"consumption"]
DF_validation_features =DF_validation.drop(["consumption","temperature","irradiance","weekOfYear"],axis=1)
predicted_consumption_validation = linear_reg.predict(DF_validation_features)
DF_comparison_LR_validation=pd.DataFrame(predicted_consumption_validation, index=DF_validation_target.index, columns=["predicted consumption_LR"])
DF_comparison_LR_validation=DF_comparison_LR_validation.join(DF_validation_target)
DF_comparison_LR_validation
```

Random Forests

```python
from sklearn.ensemble import RandomForestRegressor
RF_reg=RandomForestRegressor()
RF_reg.fit(DF_training_features, DF_training_target)
```

```python
predicted_consumption_Validation_RF=RF_reg.predict(DF_validation_features)
DF_comparison_RF_validation=pd.DataFrame(predicted_consumption_Validation_RF, index=DF_validation_target.index, columns=["PredictedConsumption_RF"])
DF_comparison_RF_validation=DF_comparison_RF_validation.join(DF_validation_target)
DF_comparison_LR_validation=pd.DataFrame(predicted_consumption_validation, index=DF_validation_target.index, columns=["predicted consumption_LR"])
DF_comparison_validation =DF_comparison_RF_validation.join(DF_comparison_LR_validation)
DF_comparison_validation
```

Errors

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
MAE_LR_validation=mean_absolute_error(predicted_consumption_validation, DF_validation_target)
MSE_LR_validation=mean_squared_error(predicted_consumption_validation, DF_validation_target)
R2_LR_validation=r2_score(predicted_consumption_validation, DF_validation_target)
print(f"MAE_LRL: {MAE_LR_validation}, MSE_LRL: {MSE_LR_validation}, R2_LRL: {R2_LR_validation}")
```

```python
MAE_RF_validation=mean_absolute_error(predicted_consumption_Validation_RF, DF_validation_target)
MSE_RF_validation=mean_squared_error(predicted_consumption_Validation_RF, DF_validation_target)
R2_RF_validation=r2_score(predicted_consumption_Validation_RF, DF_validation_target)
print(f"MAE_RF: {MAE_RF_validation}, MSE_RF: {MSE_RF_validation}, R2_RF: {R2_RF_validation}")
```
