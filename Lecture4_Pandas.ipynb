```python
import pandas as pd
```

```python
resistances_names = ["R1", "R2", "R3", "R4", "R5"]
names_ofParameters = ["type", "h", "k", "L", "RValue"]
resistances_types = ["conv", "cond", "cond", "cond", "conv"]
resistances_h = [10, None, None, None, 40]
resistances_k = [None, 0.8, 1.5, 0.05, None]
resistances_L = [None, 0.2, 0.3, 0.6, None]
resistances_RValue = [0,0,0,0,0]
resistances_listOfLists = [resistances_types, resistances_h, resistances_k, resistances_L, resistances_RValue]
resistances_DF = pd.DataFrame(resistances_listOfLists, index=names_ofParameters, columns=resistances_names)
resistances_DF = resistances_DF.transpose()
resistances_DF
```

```python
isCond = resistances_DF.loc[:, "type"] == "cond"
resistances_DF.loc[isCond, "RValue"] = resistances_DF.loc[isCond, "L"]/resistances_DF.loc[isCond, "k"]
resistances_DF
```

```python
isConv = resistances_DF.loc[:, "type"] == "conv"
resistances_DF.loc[isConv, "RValue"] = 1/resistances_DF.loc[isConv, "h"]
resistances_DF
```
