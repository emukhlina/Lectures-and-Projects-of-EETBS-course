## It's impossible to compare all days of the year, we need to define accuracy metrics to evaluate in a quantitative manner, the performance of the two models

Some accuracy metrics
* Mean absolute error
* Mean squared error
* R2 Score

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
MAE_LR_validation=mean_absolute_error(predicted_consumption_validation, DF_validation_target)
MSE_LR_validation=mean_squared_error(predicted_consumption_validation, DF_validation_target)
R2_LRL_validation=r2_score(predicted_consumption_validation, DF_validation_target)
print(f"MAE_LRL: {MAE_LR_validation}, MSE_LRL: {MSE_LR_validation}, R2_LRL: {R2_LRL_validation}")
```

We broke down the data randomly. Which would introduce randomness to our results.

```python
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import cross_val_score
DF_trainingandValidation_features=DF_trainingandValidation.drop(["consumption", "temperature", "irradiance"], axis=1)
DF_trainingandValidation_target=DF_trainingandValidation.loc[:,"consumption"]
```

```python
predicted_consumption_CV_LR=cross_val_predict(linear_reg, DF_trainingandValidation_features, DF_trainingandValidation_target, cv=10)
predicted_consumption_CV_RF=cross_val_predict(RF_reg, DF_trainingandValidation_features, DF_trainingandValidation_target, cv=10)
```

```python
R2_LR_validation_Crossval=r2_score(predicted_consumption_CV_LR, DF_trainingandValidation_target)
R2_RF_validation_Crossval=r2_score(predicted_consumption_CV_RF, DF_trainingandValidation_target)
```
